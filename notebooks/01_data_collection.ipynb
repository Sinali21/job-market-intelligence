```python
# ── IMPORTS ──────────────────────────────────
# We load all the tools we need at the start

import pandas as pd        # For creating and managing data tables
import numpy as np         # For number operations
import random              # For generating random values
import os                  # For file and folder operations
from datetime import datetime  # For getting today's date
from tqdm.notebook import tqdm # For showing progress bars

print("✅ All libraries imported successfully!")
```

```python
# ── PROJECT PATHS ────────────────────────────
# We define all folder paths in one place
# So if we move the project, we only change it here

BASE_PATH = 'D:\Sinali\projects\job_market_intelligence_platform'

RAW_DATA_PATH      = f'{BASE_PATH}/data/raw'
PROCESSED_DATA_PATH = f'{BASE_PATH}/data/processed'
OUTPUT_PATH        = f'{BASE_PATH}/outputs'

# Create folders if they don't exist
os.makedirs(RAW_DATA_PATH, exist_ok=True)
os.makedirs(PROCESSED_DATA_PATH, exist_ok=True)
os.makedirs(OUTPUT_PATH, exist_ok=True)

print("📁 Project paths set:")
print(f"   Raw data      → {RAW_DATA_PATH}")
print(f"   Processed data→ {PROCESSED_DATA_PATH}")
print(f"   Outputs       → {OUTPUT_PATH}")
print("\n✅ All folders confirmed!")
```

```python
# ── LOAD DATASET ─────────────────────────────
# We load the CSV file into a pandas DataFrame
# A DataFrame is like an Excel table in Python

df = pd.read_csv('D:/Sinali/projects/job_market_intelligence_platform/data/raw/job_descriptions.csv')

print(f"✅ Dataset loaded!")
print(f"📊 Total rows    : {len(df)}")
print(f"📊 Total columns : {len(df.columns)}")
print(f"\n📋 Column names:")
for col in df.columns:
    print(f"   → {col}")
```

```python
# ── PREVIEW DATA ─────────────────────────────
# Let's see the first 3 rows to understand the data

print("📋 FIRST 3 ROWS:")
df.head(3)
```

```python
# ── CHECK MISSING VALUES ─────────────────────
# Missing values = empty cells in our dataset
# We need to know which columns have gaps

print("📋 MISSING VALUES PER COLUMN:")
print("=" * 40)

missing = df.isnull().sum()
percent = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count' : missing,
    'Missing %'     : percent.round(2)
})

print(missing_df)
```

```python
# ── SELECT RELEVANT COLUMNS ──────────────────
# From 23 columns we keep only what we need
# This makes our dataset lighter and faster to process

df_selected = df[[
    'Job Title',
    'Job Description',
    'skills',
    'Salary Range',
    'Experience',
    'Qualifications',
    'location',
    'Country',
    'Work Type',
    'Company',
    'Job Posting Date',
    'Role'
]]

# Rename columns to simpler names
df_selected.columns = [
    'title',
    'description',
    'skills',
    'salary',
    'experience',
    'qualification',
    'location',
    'country',
    'work_type',
    'company',
    'date_posted',
    'role'
]

print(f"✅ Selected {len(df_selected.columns)} columns")
print(f"📊 Shape: {df_selected.shape}")
print(f"\n📋 Sample:")
df_selected.head(3)
```

```python
# ── SAVE SELECTED DATA ───────────────────────
# We save this cleaned selection as our working dataset
# This is what all other notebooks will use

save_path = f'{RAW_DATA_PATH}/jobs_raw.csv'

df_selected.to_csv(save_path, index=False)

print(f"✅ Dataset saved!")
print(f"📁 Location : {save_path}")
print(f"📊 Rows     : {len(df_selected)}")
print(f"📊 Columns  : {len(df_selected.columns)}")
print(f"\n🎉 Notebook 1 Complete!")
print(f"👉 Next step → Open Notebook 2: Data Preprocessing")
```
